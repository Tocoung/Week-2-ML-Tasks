# Random Forests and Decision Trees
- Classify the Spotify.csv dataset 
- Predict the Genre 
- The dataset has 23 Columns and The Output feature is **playlist_genre** so all others can be used as input features. Decide on what can be used and what cannot be used. 
- Do the Preprocessing below
- Do 70:30 Train test split 

```python
import pandas as pd 
data = pd.read_csv('C:\\Users\\prith\\ML-fundamentals\\Week 2\\RF and DT Assignment\\spotify Edited.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
columns_to_drop = ["track_id", 'track_name', 'track_artist', 'track_album_id', 'track_album_name', 'track_album_release_date', 'playlist_name', 'playlist_id']
present_columns_to_drop = [col for col in columns_to_drop if col in data.columns]
data = data.drop(columns=present_columns_to_drop)
data.info()
```

```python
data.head()
```

```python
label_encoder = LabelEncoder()
data['playlist_genre'] = label_encoder.fit_transform(data['playlist_genre'])
data['playlist_subgenre'] = label_encoder.fit_transform(data['playlist_subgenre'])
data.head()
```

```python
X = data.drop(columns=['playlist_genre'])
y = data['playlist_genre']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Frame a Decision Tree Model 
- Use Any Parameters
- You can use any library

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, y_pred_dt)
dt_classification_report = classification_report(y_test, y_pred_dt)
dt_accuracy
```

```python
print(dt_classification_report)
```

## Frame a Random Forests Model 
- Use Any Parameters
- You can use any library

```python
from sklearn.ensemble import RandomForestClassifier


rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predict using the Random Forest model
y_pred_rf = rf_model.predict(X_test)

# Evaluate the Random Forest model
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_classification_report = classification_report(y_test, y_pred_rf)
rf_accuracy
```

```python
print(rf_classification_report)
```
